```python
#cut of polynomial at given total degree
def cut(f,n):
    g = 0
    for i in range(n+1):
        for j in range(i+1):
            g = g + f.coefficient({y:j,x:(i-j)})*y^j*x^(i-j)
    return g

#implement algo from Schneiders book computes the grouplaw up to total degree deg for a given prime p and frobenius polynomial f
def grouplaw(f,deg,p):
    Polys = [0,x+y]
    for i in range(1,deg):
        H = Polys[i]
        F = f(Polys[i])-Polys[i](f(x),f(y))
        tG = cut(F,i+1)
        G = H + (-1/(p-p^(i+1)))*tG
        Polys.append(G)
    return(Polys[deg])
```
