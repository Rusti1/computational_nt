```python
#cut of polynomial at given total degree
def cut(f,n):
    g = 0
    for i in range(n+1):
        for j in range(i+1):
            g = g + f.coefficient({y:j,x:(i-j)})*y^j*x^(i-j)
    return g

#implement algo from Schneiders book computes the grouplaw up to total degree deg for a given prime p and frobenius polynomial f
def grouplaw(f,deg,p):
    Polys = [0,x+y]
    for i in range(1,deg):
        H = Polys[i]
        F = f(Polys[i])-Polys[i](f(x),f(y))
        tG = cut(F,i+1)
        G = H + (-1/(p-p^(i+1)))*tG
        Polys.append(G)
    return(Polys[deg])
```

```python
#Fast X+u following Berger. 
def cutuni(n,f):
    q = f.variables()[0]
    f += O(q^n)   # truncating as a power series
    f = f.polynomial()
    return f

def fasttrans(n,der,u,p,f,phi):
    coeffs = [u]
    h = coeffs[0]
    c = int(p-1)
    t = phi.variables()[0]
    for i in range(1,n):
        rest = cutuni(i,phi(h)-phi)
        co = -rest/der
        coeffs.append(co)
        h = h + coeffs[i]*t^i
    return h

        
        
    


















```
